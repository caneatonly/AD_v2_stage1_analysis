# CFD 仿真结果数据审查与验证报告

## 审查目标

本报告用于对 `static_aoa_data` 中的 static AoA CFD 数据进行物理合理性审查，并为论文 Section 4.2 的 `C_X(lpha), C_Z(lpha), C_m(lpha)` 证据链提供诊断依据。

审查流程沿用“五步验证法”：

1. **曲线形态检查 (Trend Validation)**：检查 $C_X, C_Z, C_m$ 在 $0^\circ$--$180^\circ$ 全攻角范围内是否连续、平滑并符合大攻角流动趋势。
2. **量级与物理极限检查 (Magnitude Check)**：校核 0° 轴向阻力和 90° 横流阻力量级。
3. **压心位置检查 (Center of Pressure, CP)**：由 $C_m/C_Z$ 反推压心位置，判断力矩是否物理合理。
4. **残差与监测曲线检查 (Residual / Monitor Pattern)**：检查关键攻角的残差和力/力矩监测曲线是否稳定。
5. **最终验证图表 (Final Validation Plots)**：生成 $0^\circ$--$180^\circ$ 全量系数图、深分离区局部图和压心图。

注意：本 notebook 只审查 static AoA CFD maps，不用于识别 added mass、damping 或完整动态 release 过程。



```python
from pathlib import Path
import glob
import os
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Publication-style plotting
plt.style.use('seaborn-v0_8-whitegrid')
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = ['Times New Roman'] + mpl.rcParams['font.serif']
mpl.rcParams['font.size'] = 12
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['xtick.labelsize'] = 11
mpl.rcParams['ytick.labelsize'] = 11
mpl.rcParams['legend.fontsize'] = 10
mpl.rcParams['figure.dpi'] = 160
mpl.rcParams['savefig.dpi'] = 300
plt.rcParams['axes.unicode_minus'] = False

# Physical / reference quantities
rho = 997.561
V_total = 1.0
A_ref = 0.0056745
L_ref = 0.625
D = np.sqrt(4 * A_ref / np.pi)
print(f"Calculated equivalent diameter D: {D:.6f} m")

# Data path: current project data folder
PROJECT_ROOT = Path.cwd()
data_dir = PROJECT_ROOT / 'static_aoa_data'
out_dir = PROJECT_ROOT / 'sim_flip' / 'results' / 'static_aoa_review' / 'notebook_figs'
out_dir.mkdir(parents=True, exist_ok=True)
print(f"Data directory: {data_dir}")
print(f"Figure output directory: {out_dir}")


def load_raw_monitor_data(data_dir: Path, n_last: int = 500) -> pd.DataFrame:
    """Load C_*_overmesh.csv files and compute last-window mean/std."""
    files = sorted(data_dir.glob('C_*_overmesh.csv'), key=lambda p: int(re.search(r'C_(\d+)_overmesh', p.name).group(1)))
    rows = []
    for file_path in files:
        match = re.search(r'C_(\d+)_overmesh\.csv', file_path.name)
        if not match:
            continue
        angle = int(match.group(1))
        df = pd.read_csv(file_path)
        required = {'Cx Monitor', 'Cz Monitor', 'Cm Monitor'}
        if not required.issubset(df.columns):
            raise ValueError(f"{file_path.name} missing columns: {required - set(df.columns)}")
        tail_len = min(len(df), n_last)
        tail = df.tail(tail_len)
        rows.append({
            'Angle': angle,
            'Cx': tail['Cx Monitor'].mean(),
            'Cz': tail['Cz Monitor'].mean(),
            'Cm': tail['Cm Monitor'].mean(),
            'Cx_std': tail['Cx Monitor'].std(ddof=1),
            'Cz_std': tail['Cz Monitor'].std(ddof=1),
            'Cm_std': tail['Cm Monitor'].std(ddof=1),
            'n_samples': len(df),
            'window': tail_len,
            'File': file_path.name,
        })
    return pd.DataFrame(rows).sort_values('Angle').reset_index(drop=True)


def load_corrected_table(data_dir: Path) -> pd.DataFrame | None:
    path = data_dir / 'Hydrodynamic_Coefficients_Corrected.csv'
    if not path.exists():
        return None
    df = pd.read_csv(path)
    required = {'Angle', 'Cx', 'Cz', 'Cm'}
    if not required.issubset(df.columns):
        raise ValueError(f"{path.name} missing columns: {required - set(df.columns)}")
    return df.sort_values('Angle').reset_index(drop=True)


def add_local_residuals(df: pd.DataFrame, cols=('Cx', 'Cz', 'Cm')) -> pd.DataFrame:
    out = df.copy()
    for col in cols:
        vals = out[col].to_numpy(float)
        resid = np.full(len(vals), np.nan)
        for i in range(1, len(vals) - 1):
            resid[i] = vals[i] - 0.5 * (vals[i - 1] + vals[i + 1])
        med = np.nanmedian(resid)
        mad = np.nanmedian(np.abs(resid - med))
        robust_z = 0.6745 * np.abs(resid - med) / mad if mad > 0 else np.full(len(vals), np.nan)
        out[f'{col}_local_resid'] = resid
        out[f'{col}_robust_z'] = robust_z
    return out


def get_local_outliers(df: pd.DataFrame, threshold: float = 3.5) -> pd.DataFrame:
    rows = []
    for _, row in df.iterrows():
        for col in ['Cx', 'Cz', 'Cm']:
            z = row.get(f'{col}_robust_z', np.nan)
            if pd.notna(z) and z >= threshold:
                rows.append({
                    'Angle': int(row['Angle']),
                    'Coefficient': col,
                    'Value': row[col],
                    'LocalResidual': row[f'{col}_local_resid'],
                    'RobustZ': z,
                })
    return pd.DataFrame(rows).sort_values('RobustZ', ascending=False).reset_index(drop=True) if rows else pd.DataFrame()


df_raw = load_raw_monitor_data(data_dir, n_last=500)
df_corrected = load_corrected_table(data_dir)

# Use corrected table as plotting/review table if it exists, but always keep raw data for comparison.
df_coeffs = df_corrected.copy() if df_corrected is not None else df_raw[['Angle', 'Cx', 'Cz', 'Cm', 'File']].copy()
df_raw_local = add_local_residuals(df_raw)
df_coeffs_local = add_local_residuals(df_coeffs)

print("Loaded raw monitor data:")
print(df_raw[['Angle', 'Cx', 'Cz', 'Cm', 'Cx_std', 'Cz_std', 'Cm_std', 'n_samples', 'window', 'File']].to_string(index=False))
print(f"\nTotal raw angles loaded: {len(df_raw)}")
print(f"Angle range: {df_raw['Angle'].min()} to {df_raw['Angle'].max()} deg")
print(f"Unique angle steps: {np.unique(np.diff(df_raw['Angle'].to_numpy()))}")

if df_corrected is not None:
    print("\nCorrected coefficient table found and will be used as df_coeffs for final curves.")
    merged = df_raw[['Angle', 'Cx', 'Cz', 'Cm']].merge(
        df_corrected[['Angle', 'Cx', 'Cz', 'Cm']], on='Angle', suffixes=('_raw', '_corrected')
    )
    for col in ['Cx', 'Cz', 'Cm']:
        merged[f'{col}_delta'] = merged[f'{col}_corrected'] - merged[f'{col}_raw']
    df_changed = merged[(merged[['Cx_delta', 'Cz_delta', 'Cm_delta']].abs() > 1e-6).any(axis=1)]
    print("\nRaw vs corrected changed points:")
    print(df_changed.to_string(index=False))
else:
    df_changed = pd.DataFrame()
    print("\nNo corrected table found; raw last-window means will be used.")


```

## 第一步：曲线形态检查 (Trend Validation)

**检查标准：**

- **$C_X$（轴向力）**：0° 附近应反映轴向阻力，大攻角区不应出现孤立非物理跳变。
- **$C_Z$（法向力）**：0° 附近应接近 0，90° 附近达到高值并在大攻角区平滑过渡。
- **$C_m$（俯仰力矩）**：应在攻角变化中呈现可解释的过零和峰值行为，局部强跳变需要额外检查。

本节同时比较 raw last-500 均值与 review/corrected table，用于标记被修正的可疑点。



```python
# Plot 0-180 deg coefficient curves: raw vs corrected
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)
series = [('Cx', '$C_X$'), ('Cz', '$C_Z$'), ('Cm', '$C_m$')]

for ax, (col, label) in zip(axes, series):
    ax.plot(df_raw['Angle'], df_raw[col], 'o--', alpha=0.45, label=f'{label} raw last-500')
    ax.plot(df_coeffs['Angle'], df_coeffs[col], 's-', linewidth=2.2, label=f'{label} review table')
    ax.axhline(0, color='black', linewidth=0.8)
    ax.axvspan(85, 125, color='tab:orange', alpha=0.08, label='85-125 deg high-risk region' if col == 'Cx' else None)
    ax.set_ylabel(label)
    ax.grid(True, alpha=0.3)
    ax.legend(loc='best')

axes[-1].set_xlabel(r'Angle of Attack $\alpha$ (deg)')
axes[-1].set_xlim(-5, 185)
axes[-1].set_xticks(np.arange(0, 181, 15))
fig.suptitle('Static AoA CFD Coefficient Maps, 0-180 deg', y=1.01)
fig.tight_layout()
fig.savefig(out_dir / 'fig_coefficients_0_180_raw_vs_review.png', bbox_inches='tight')
plt.show()

print("Values in high-risk region 75-135 deg:")
print(df_coeffs[(df_coeffs['Angle'] >= 75) & (df_coeffs['Angle'] <= 135)][['Angle', 'Cx', 'Cz', 'Cm']].to_string(index=False))

print("\nRaw local outliers (threshold robust-z >= 3.5):")
print(get_local_outliers(df_raw_local, threshold=3.5).to_string(index=False))
print("\nReview-table local outliers (threshold robust-z >= 3.5):")
print(get_local_outliers(df_coeffs_local, threshold=3.5).to_string(index=False))


```

## 第二步：量级与物理极限检查 (Magnitude Check)

### 1. 0° 工况：轴向阻力校核

- 参考判据：短粗微型水下机体，0° $C_X$ 常见量级可先以 -0.2 到 -0.4 作为初步检查范围。

### 2. 90° 工况：横流阻力校核

- 使用等效直径 $D$ 和投影面积 $LD$ 将基于 $A_{ref}$ 的 $C_Z$ 换算为横流 $C_D$。
- 若换算值偏低或偏高，优先复核坐标系、force report 方向、参考面积和投影面积定义。



```python
# 1. 0 deg axial drag check
cx_0 = df_coeffs.loc[df_coeffs['Angle'] == 0, 'Cx'].values[0] if 0 in df_coeffs['Angle'].values else np.nan
print(f"--- 0 Degree Drag Check ---")
print(f"Cx at 0 deg: {cx_0:.6f}")
if -0.4 <= cx_0 <= -0.2:
    print("Result: PASS (within the preliminary range -0.2 to -0.4)")
else:
    print("Result: WARNING (outside the preliminary range; check geometry/reference area)")

# 2. 90 deg cross-flow drag check
cz_90 = df_coeffs.loc[df_coeffs['Angle'] == 90, 'Cz'].values[0] if 90 in df_coeffs['Angle'].values else np.nan
if not np.isnan(cz_90):
    conversion_factor = (np.pi * D) / (4 * L_ref)  # A_ref / projected area for circular A_ref
    cd_cross = abs(cz_90) * conversion_factor
    print(f"\n--- 90 Degree Cross-flow Drag Check ---")
    print(f"Cz at 90 deg based on A_ref: {cz_90:.6f}")
    print(f"Projected area L*D: {L_ref * D:.6f} m^2")
    print(f"Reference area A_ref: {A_ref:.6f} m^2")
    print(f"Conversion factor A_ref/A_proj: {conversion_factor:.6f}")
    print(f"Calculated Cd_cross: {cd_cross:.6f}")
    if 0.8 <= cd_cross <= 1.3:
        print("Result: PASS against the simple cylinder-level range 0.8-1.3")
    elif cd_cross < 0.8:
        print("Result: WARNING: low cross-flow Cd; re-check force direction, reference area, projected area, and body-frame definition")
    else:
        print("Result: HIGH DRAG: plausible only if fairing/open-mouth geometry strongly increases cross-flow drag")
else:
    print("\nWARNING: 90 degree data not found.")

# 3. Symmetry/end-point sanity checks
cx_180 = df_coeffs.loc[df_coeffs['Angle'] == 180, 'Cx'].values[0] if 180 in df_coeffs['Angle'].values else np.nan
cz_180 = df_coeffs.loc[df_coeffs['Angle'] == 180, 'Cz'].values[0] if 180 in df_coeffs['Angle'].values else np.nan
cm_180 = df_coeffs.loc[df_coeffs['Angle'] == 180, 'Cm'].values[0] if 180 in df_coeffs['Angle'].values else np.nan
print(f"\n--- 180 Degree End-point Check ---")
print(f"Cx(180)={cx_180:.6f}, Cz(180)={cz_180:.6f}, Cm(180)={cm_180:.6f}")
print("Interpretation: Cz and Cm near zero at 180 deg are expected; Cx sign/magnitude depends on the body-axis force convention.")


```

## 第三步：压心变化图 (Center of Pressure, CP)

压心位置按下式计算：

$$
X_{cp}(lpha) pprox -rac{C_m(lpha)}{C_Z(lpha)} L_{ref}
$$

当 $|C_Z|$ 过小时，$X_{cp}$ 对噪声非常敏感，因此低攻角点只作定性参考。



```python
# Compute Xcp and filter points with very small |Cz|
cz_eps = 0.01
df_cp = df_coeffs[abs(df_coeffs['Cz']) > cz_eps].copy()
df_cp['Xcp'] = - (df_cp['Cm'] / df_cp['Cz']) * L_ref

fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(df_cp['Angle'], df_cp['Xcp'], 'd-', color='#6a3d9a', label='$X_{cp}$')
ax.axhline(0.325, color='gray', linestyle='--', alpha=0.7, label='Body nose (+0.325 m)')
ax.axhline(-0.300, color='gray', linestyle='--', alpha=0.7, label='Body tail (-0.300 m)')
ax.axhline(0, color='black', linewidth=0.8, label='CG / origin')
ax.axvspan(85, 125, color='tab:orange', alpha=0.08, label='85-125 deg high-risk region')
ax.set_xlabel(r'Angle of Attack $\alpha$ (deg)')
ax.set_ylabel(r'$X_{cp}$ Position (m)')
ax.set_xlim(-5, 185)
ax.set_xticks(np.arange(0, 181, 15))
ax.grid(True, alpha=0.3)
ax.legend(loc='best')
fig.tight_layout()
fig.savefig(out_dir / 'fig_xcp_0_180.png', bbox_inches='tight')
plt.show()

print("Xcp summary:")
print(df_cp['Xcp'].describe())
print("\nXcp values at key angles:")
key_angles = [5, 10, 45, 85, 90, 100, 125, 135, 170, 175]
print(df_cp[df_cp['Angle'].isin(key_angles)][['Angle', 'Xcp']].to_string(index=False))

outside = df_cp[(df_cp['Xcp'] > 0.325) | (df_cp['Xcp'] < -0.300)]
print("\nXcp outside nominal body range:")
print(outside[['Angle', 'Xcp']].to_string(index=False) if len(outside) else 'None')


```

## 第四步：残差与监测曲线检查 (Residual / Monitor Pattern)

检查方法：读取 `Residuals_{angle}.csv` 并检查关键攻角的残差曲线。

- 合理：残差下降后进入小幅震荡，力/力矩监测曲线无长期漂移。
- 不合理：残差发散，或者力/力矩均值持续单向漂移。

本节优先展示 0°、45°、85°、90°、125°、180° 等代表性工况。



```python
def plot_residuals(angles_to_check):
    fig, axes = plt.subplots(1, len(angles_to_check), figsize=(4.4 * len(angles_to_check), 4), squeeze=False)
    axes = axes[0]
    for ax, angle in zip(axes, angles_to_check):
        res_file = data_dir / f'Residuals_{angle}.csv'
        if res_file.exists():
            try:
                df_res = pd.read_csv(res_file)
                x_col = df_res.columns[0]
                for col in df_res.columns[1:5]:
                    ax.semilogy(df_res[x_col], df_res[col], label=col, linewidth=0.8)
                ax.set_title(f'Residuals at {angle} deg')
                ax.set_xlabel('Iteration')
                ax.grid(True, which='both', alpha=0.25)
                ax.legend(fontsize=7)
            except Exception as e:
                ax.set_title(f'{angle} deg read error')
                ax.text(0.05, 0.5, str(e), transform=ax.transAxes, fontsize=8)
        else:
            ax.set_title(f'{angle} deg missing')
            ax.text(0.05, 0.5, f'Missing {res_file.name}', transform=ax.transAxes)
    axes[0].set_ylabel('Residual')
    fig.tight_layout()
    fig.savefig(out_dir / 'fig_residuals_key_angles.png', bbox_inches='tight')
    plt.show()

# Show representative low, mid, suspicious, and endpoint angles
plot_residuals([0, 45, 85, 90, 125, 180])

# Flag high standard deviation in the last 500 samples
stability_flags = []
for _, row in df_raw.iterrows():
    flags = []
    if row['Cx_std'] > 0.05:
        flags.append('Cx_std_high')
    if row['Cz_std'] > 0.20:
        flags.append('Cz_std_high')
    if row['Cm_std'] > 0.05:
        flags.append('Cm_std_high')
    if flags:
        stability_flags.append({'Angle': int(row['Angle']), 'Flags': ';'.join(flags)})

df_stability_flags = pd.DataFrame(stability_flags)
print("Monitor stability flags:")
print(df_stability_flags.to_string(index=False) if len(df_stability_flags) else 'None under current thresholds')


```

## 第五步：最终验证图表 (Final Validation Plots)

生成 3 张用于 4.2 初步审查的关键图表：

1. **全攻角力系数图**：$0^\circ$--$180^\circ$ 的 $C_X, C_Z, C_m$，同时显示 raw 与 review/corrected table。
2. **深分离区局部放大图**：$75^\circ$--$135^\circ$，重点检查 85°、90°、120°、125°、130°。
3. **压心变化图**：$X_{cp}$ 随攻角变化，检查力矩和法向力组合是否物理合理。



```python
# Figure 1: full 0-180 coefficient maps, one panel per coefficient
fig1, axes = plt.subplots(3, 1, figsize=(10, 10), sharex=True)
for ax, (col, ylabel, color) in zip(
    axes,
    [('Cx', '$C_X$', '#1f77b4'), ('Cz', '$C_Z$', '#2ca02c'), ('Cm', '$C_m$', '#d62728')]
):
    ax.plot(df_raw['Angle'], df_raw[col], 'o--', color=color, alpha=0.35, label='raw last-500')
    ax.plot(df_coeffs['Angle'], df_coeffs[col], 's-', color=color, linewidth=2.2, label='review table')
    ax.axhline(0, color='black', linewidth=0.8)
    ax.axvspan(85, 125, color='tab:orange', alpha=0.08)
    ax.set_ylabel(ylabel, fontweight='bold')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend(loc='best')
axes[-1].set_xlabel(r'Angle of Attack $\alpha$ (deg)', fontweight='bold')
axes[-1].set_xlim(-5, 185)
axes[-1].set_xticks(np.arange(0, 181, 15))
fig1.tight_layout()
fig1.savefig(out_dir / 'paper_draft_coefficients_0_180.png', bbox_inches='tight')
plt.show()

# Figure 2: zoomed high-risk separation region
fig2, ax2 = plt.subplots(figsize=(9, 5))
df_zoom = df_coeffs[(df_coeffs['Angle'] >= 75) & (df_coeffs['Angle'] <= 135)]
for col, label, color, marker in [
    ('Cx', '$C_X$', '#1f77b4', 'o'),
    ('Cz', '$C_Z$', '#2ca02c', 's'),
    ('Cm', '$C_m$', '#d62728', '^'),
]:
    ax2.plot(df_zoom['Angle'], df_zoom[col], marker + '-', color=color, label=label, markeredgecolor='white')
ax2.axhline(0, color='black', linewidth=0.8)
for a in [85, 90, 120, 125, 130]:
    ax2.axvline(a, color='gray', linestyle=':', linewidth=0.8, alpha=0.6)
ax2.set_xlabel(r'Angle of Attack $\alpha$ (deg)', fontweight='bold')
ax2.set_ylabel('Coefficient', fontweight='bold')
ax2.set_xticks(np.arange(75, 136, 5))
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend(frameon=True)
fig2.tight_layout()
fig2.savefig(out_dir / 'paper_draft_coefficients_zoom_75_135.png', bbox_inches='tight')
plt.show()

# Figure 3: center of pressure, full 0-180
fig3, ax3 = plt.subplots(figsize=(10, 5))
ax3.plot(df_cp['Angle'], df_cp['Xcp'], 'D-', color='#9467bd', label='$X_{cp}$', markeredgecolor='white')
ax3.axhline(0, color='black', linestyle='-', linewidth=1.0, label='CG / origin')
ax3.axhline(0.325, color='gray', linestyle='--', alpha=0.7, label='Body nose (+0.325 m)')
ax3.axhline(-0.300, color='gray', linestyle='--', alpha=0.7, label='Body tail (-0.300 m)')
ax3.axvspan(85, 125, color='tab:orange', alpha=0.08, label='85-125 deg high-risk region')
ax3.set_xlabel(r'Angle of Attack $\alpha$ (deg)', fontweight='bold')
ax3.set_ylabel(r'Center of Pressure $X_{cp}$ (m)', fontweight='bold')
ax3.set_xlim(-5, 185)
ax3.set_xticks(np.arange(0, 181, 15))
ax3.grid(True, linestyle='--', alpha=0.5)
ax3.legend(frameon=True, fontsize=9)
fig3.tight_layout()
fig3.savefig(out_dir / 'paper_draft_xcp_0_180.png', bbox_inches='tight')
plt.show()


```
